### What is the structure of your dataset?

The Prosper Loan Dataset has 113,937 loan records and 81 features in total. These features gives essential information about the loan as well as the borrower's essential information.

### What is/are the main feature(s) of interest in your dataset?

After evaluation, I considered the following features important for my analysis:
1. `LoanKey` - Unique key for each loan.
2. `CreditGrade` -The Credit rating that was assigned at the time the listing went live.
3. `IncomeRange` - The income range of the borrower at the time the listing was created.
4. `StatedMonthlyIncome` - The monthly income the borrower stated at the time the listing was created.
5. `LoanOriginalAmount` - The origination amount of the loan.
6. `Term` - The length of the loan expressed in months.
7. `LoanStatus` - The current status of the loan: Cancelled,  Chargedoff, Completed, Current, Defaulted, FinalPaymentInProgress, PastDue.
8. `BorrowerAPR` - The Borrower's Annual Percentage Rate (APR) for the loan.
9. `BorrowerRate` - The Borrower's interest rate for this loan. 
10. `ProsperScore` - A custom risk score built using historical Prosper data. The score ranges from 1-10, with 10 being the best, or lowest risk score.
11. `TotalProsperLoans` - Number of Prosper loans the borrower at the time they created this listing. This value will be null if the borrower had no prior loans. 
12. `EmploymentStatus` - The employment status of the borrower at the time they posted the listing.
13. `DebtToIncomeRatio` - The debt to income ratio of the borrower at the time the credit profile was pulled. This value is Null if the debt to income ratio is not available.
14. `EstimatedReturn` - The estimated return assigned to the listing at the time it was created.
15. `EstimatedLoss` - Estimated loss is the estimated principal loss on charge-offs.
16. `Investors` - The number of investors that funded the loan.

#### __Question 10__: How does a borrower's `IncomeRange` & `CreditGrade` affect their loans' interest rates(`BorrowerRate`)?

```python
# Checkunique Qualitative Values of Credit Grade
print(f"Unique Credit Grade Values: {loan_df.CreditGrade.unique()}")
print()
print(f"Unique Income Range Values: {loan_df.IncomeRange.unique()}")
```

```python
# Convert Listings with no salary to 0
loan_df["IncomeRange"] = loan_df.IncomeRange.replace(['Not displayed'], '$0')
```

```python
# DATATYPE CONVERSION string to ordered categorical

# Income Range
# Convert Income Ranges to categorical datatype
income_ranges = ['Not employed',  '$0', '$1-24,999', '$25,000-49,999', '$50,000-74,999', '$75,000-99,999', '$100,000+']

# Returns Income Ranges as a categorical datatype
salary_range = pd.api.types.CategoricalDtype(ordered=True, categories=income_ranges)

# Use pandas.astype() to convert the "IncomeRange" column from a plain object type into an ordered categorical type 
loan_df['IncomeRange'] = loan_df['IncomeRange'].astype(salary_range);


# CreditGrade
# Convert Credit Grade to categorical datatype
credit_types = ['HR',  'E', 'D', 'C', 'B', 'A', 'AA']

# Returns CreditGrades as a categorical datatype
credit_class = pd.api.types.CategoricalDtype(ordered=True, categories=credit_types)

# Use pandas.astype() to convert the "CreditGrade" column from a plain object type into an ordered categorical type 
loan_df['CreditGrade'] = loan_df['CreditGrade'].astype(credit_class);
```

```python
# Confirming datatype changes
loan_df.info()
```

##### __Visualisation__

```python
# Initialise Two Plots In One Figure
plt.figure(figsize = [24, 10])
base_color = sb.color_palette()[2]

# LEFT plot: Box Plot for Income Range classes vs BorrowRate
plt.subplot(1, 2, 1)
#Return the axes object
ax1 = sb.boxplot(data=loan_df, x='IncomeRange', y='BorrowerRate', color=base_color)
plt.xticks(rotation=45);

# RIGHT plot: Box Plot for Credit Grade classes vs BorrowRate
plt.subplot(1, 2, 2)
sb.boxplot(data=loan_df, x='CreditGrade', y='BorrowerRate', color=base_color)
plt.xticks(rotation=15);
plt.ylim(ax1.get_ylim())
```

##### __Observation__
##### 1. BorrowerRate vs IncomeRange
A general trend noticed from the visualisaton to the top left is that The median BorrowerRate(Interest Rate) decreases as a borrower's Income Range increases. Loan Borrowers under the `IncomeRange` category, 'Not Employed' have the highest `BorrowerRate` median while this value keeps on decreasing across Income Ranges __'$1-24,999' to '$100,000+'__.   The '$0' `IncomeRange `category has outliers beyond the maximum value which should be handled before explanatory analysis. The inter-quartile ranges across the `IncomeRange` categories

##### 2. BorrowerRate vs CreditGrade.
The most glaring obeservation from the top right visualisation are the tremendous amounts of outliers present across `CreditGrade` categories. These outliers are depicted by the linear data points that extend above and below the maximum and minimum values for each category.

Another obeservation are the decreasing inter-quartile ranges as CreditGrades move to higher classes (HR - AA). The Median values of `BorrowerRate` also decrease from Class HR through Class AA. This could mean that the Borrower Interest Rates value are usually lower for higher ranked `CreditGrade` classes.
